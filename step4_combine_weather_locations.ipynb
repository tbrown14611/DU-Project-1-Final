```python
# package imports go here
import pandas as pd
import numpy as np
import fastparquet as fp
```

```python
# Read TmAX weather paqquet file into a pandas dataframe to verify data integrity is maintained
weather_data_TMAX_df = pd.read_parquet('result_files/stp3_us_ca_daily_TMAX_weather_bins.parquet.gzip', engine="fastparquet") 
weather_data_TMAX_df
```

```python
# Read Ground Station locations
ground_stations_US_CA = pd.read_parquet('result_files/stp2_US_CA_Station_locations.parquet.gzip', engine="fastparquet") 

ground_stations_US_CA
```

```python
# Read in Precipitation data
TMAX_PRCP_counts = pd.read_parquet('result_files/stp3_us_ca_daily_PRCP_weather_sums.parquet.gzip', engine="fastparquet") 
TMAX_PRCP_counts
```

```python
# Combine Location and Precipitation

US_CA_loc_prcp = pd.merge(ground_stations_US_CA, TMAX_PRCP_counts, on='ID', how='inner')
US_CA_loc_prcp
```

```python
US_CA_Ground_Stations_n_locations = pd.merge(US_CA_loc_prcp, weather_data_TMAX_df, on='ID', how='inner')
US_CA_Ground_Stations_n_locations
```
